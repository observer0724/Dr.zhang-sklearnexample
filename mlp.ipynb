```python
"""Load IRIS dataset, split into train and test sets"""
iris=datasets.load_iris()
X=iris.data
encoder=OneHotEncoder()
encoder.fit(np.unique(iris.target)[:,None])
Y=encoder.transform(iris.target[:,None]).toarray()

X_tr,X_te,Y_tr,Y_te=train_test_split(X,Y,test_size=0.5)
```

```python
"""Normalize data to [-1,1]"""
scaler=MinMaxScaler(feature_range=[-1,1],copy=False)
X_tr=scaler.fit_transform(X_tr)
X_te=scaler.transform(X_te)
```

```python
"""initialize MLP classifier"""
hnode_num=10*X_tr.shape[1]

clf=MLPClassifier(hidden_layer_sizes=hnode_num,#changed
                  activation='relu',
                  solver='adam',
                  alpha=0.0001, 
                  batch_size='auto', 
                  learning_rate='constant', 
                  learning_rate_init=0.001, 
                  power_t=0.5, 
                  max_iter=2000, 
                  shuffle=True, 
                  random_state=None, 
                  tol=0.0001, 
                  verbose=True,#changed 
                  warm_start=False, 
                  momentum=0.9, 
                  nesterovs_momentum=True, 
                  early_stopping=False, 
                  validation_fraction=0.1, 
                  beta_1=0.9, 
                  beta_2=0.999, 
                  epsilon=1e-08)
```

```python
clf.fit(X_tr,Y_tr)
Y_pred=clf.predict(X_te).argmax(1)
```

```python
err=Y_pred-Y_te.argmax(1)
acc=err[err==0].shape[0]/Y_pred.shape[0]
print('Accuracy: '+str(acc))
```
